# Body Performance Categorization Project

```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from sklearn import metrics, model_selection, svm
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier

%matplotlib inline
```

### Initial Data Analysis and Processing

```python
# Looking at imported .csv
performance_data = pd.read_csv('bodyPerformance.csv')
performance_data
```

```python
# Changing gender to be numerical with 1 as male and 0 as female
performance_data.gender = performance_data.gender.astype(str).replace('M', '1').replace('F', '0').astype(float)

# Displaying statistics of data
performance_data.describe()
```

```python
# Creating figure and setting size
fig, ax = plt.subplots()
fig.set_size_inches(8, 6)

# Initializing variables for boxplot
boxplot_data = [performance_data['age'], performance_data['gender'], performance_data['height_cm'], performance_data['weight_kg'], performance_data['body fat_%'], performance_data['diastolic'], performance_data['systolic'], performance_data['gripForce'], performance_data['sit and bend forward_cm'], performance_data['sit-ups counts'], performance_data['broad jump_cm']]
flierprops = dict(marker='.', markersize=3, linestyle='none')
facecolors = dict(facecolor = ["lightblue", "darkblue"])
column=['age', 'M/F', 'wt', 'ht', 'fat', 'dia', 'sys', 'grip', 'forw', 'ups', 'jump']

# Setting up boxplot
box_plot = ax.boxplot(boxplot_data, flierprops = flierprops, labels=column, patch_artist=True);
ax.set_title('Body Performance Categorical Data')
ax.set_xlabel('Performance Categories')
ax.set_ylabel('Variable Ranges')

# Setting boxplot colors
colors = [(1, .6, .6), (1, 0.89, 0.8), (1, 0.8, 0.6), (1, 1, 0.6), (0.8, 1, 0.6), (0.6, 1, 0.6), (0.6, 1, 1), (0.6, 0.8, 1), (0.6, 0.6, 1), (1, 0.6, 1), (1, 0.6, 0.8)]
for patch, color in zip(box_plot['boxes'], colors):
    patch.set_facecolor(color)
```

```python
# Histogram of each class with sit-up ability
fig, ax = plt.subplots()
fig.set_size_inches(8, 6)
sns.histplot(performance_data, x = 'sit-ups counts', hue='class', kde=True, bins =30);
```

### Non-Descriptive Method Implementation

```python
# Scaling data with standard scalar
std_scaler = StandardScaler()
nppd = performance_data.to_numpy()
nppd
column_names=[
   'age','gender','height_cm','weight_kg', 'body fat_%', 'diastolic', 'systolic', 'gripForce', 'sit and bend forward_cm', 'sit-ups counts', 'broad jump_cm']
performance_data_scaled = std_scaler.fit_transform(nppd[:, 0:11])
performance_data_scaled = pd.DataFrame(performance_data_scaled, columns=column_names)
```

```python
# Seperating data from column to predict
y_data = performance_data.values[:, 11]
x_data = performance_data_scaled

# Creating training and testing data
x_train, x_test, y_train, y_test = model_selection.train_test_split(x_data, y_data, test_size = 0.2, random_state = 9)
```

```python
# Creating model
model = MLPClassifier(random_state=9, max_iter=5000, hidden_layer_sizes=(11,))

# Fitting data to chosen model
model.fit(x_train, y_train)

# Predicting using test data with fitted model
y_pred = model.predict(x_test)
```

```python
# Showing testing data accuracy
data_accuracy = metrics.accuracy_score(y_test, y_pred)
data_accuracy
```

```python
# Creating confusion matrix
metrics.ConfusionMatrixDisplay.from_estimator(model, x_test, y_test);
```

### User Prediction

```python
# User input for model
age = float(input("Age (20 to 65): "))
gender = float(input("Gender (1 for Male, 0 for Female): "))
height = float(input("Height in inches (37 to 84): ")) * 0.394
weight = float(input("Weight in lbs (57 to 400): ")) * 2.2
fat_percentage = float(input("Body fat percentage: (3-80) "))
diastolic = float(input("Diastolic blood pressure in mm Hg: (0-160) "))
systolic = float(input("Systolic blood pressure in mm Hg: (0-205) "))
grip_force = float(input("Grip force in lbs: (0-71) ")) * 2.2
forward = float(input("Sit and bend forward in inches: (-10-84) ")) * 0.394
sit_ups = float(input("Sit ups counts: (0-100) "))
broad_jump = float(input("Broad jump in inches: (0-119) ")) * 0.394
```

```python
# Creating dataframe with user input
user_columns = ['age','gender','height_cm','weight_kg', 'body fat_%', 'diastolic', 'systolic', 'gripForce', 'sit and bend forward_cm', 'sit-ups counts', 'broad jump_cm']
user_data = [age, gender, height, weight, fat_percentage, diastolic, systolic, grip_force, forward, sit_ups, broad_jump]
user_frame = pd.DataFrame([user_data], columns = user_columns)

# Using user dataframe to predict the performance class of the user
user_y = model.predict(user_frame)

# Displaying resulting class
fig, ax = plt.subplots()
fig.set_size_inches(1, 2)
message = f"Your performance category is of type: {user_y[0]}."
fig.text(0, 0.5, message, style = 'italic', fontsize = 30, color = "orange");
matplotlib.rc('axes',edgecolor='w')
plt.tick_params(left = False, right = False, labelleft = False, labelbottom = False, bottom = False)
```
